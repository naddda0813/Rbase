### 기본 통계량으로 분석해보기

```r
# 신약실험 
#밀가루 사용 
flour <- c(3,-2,-1,0,1,-2)
#다이어트 약 
diet <- c(-4,1,-3,-5,-2,-8)
```

```r
total <- c(flour, diet)
total
```

```r
#기본 통계량
summary(total)
```

```r
#히스토그램 
hist(total)
```

```r
boxplot(total)
```

```r
boxplot(flour,diet)
```

```r
# 상관게수 없음
cor(flour,diet)
```

```r
 x <- c(1,3,9) 
 y <- c(10,20,30) 
```

```r
par(mfrow=c(1,2))
plot(x,type="l")
plot(y,type="l")
```

```r
cor(x,y)        #벡터의 순서가 바뀔경우 상관계수가 바뀜 
```

# 커피 그래프 만들기


```r
cafe<- read.csv("../Data/coffee.csv")
head(cafe)
```

```r
str(cafe)
```

```r
summary(cafe)
```

```r
table(cafe$Coffee)
```

```r
# 범위 빈도수 
coffee_Range <- seq(0,60,10)        #범위 
table(cut(cafe$Coffee,breaks = coffee_Range,right=F))
```

```r
# 변동계수 
#coffee.cv <- coffee.sd / coffee.mean
coffee.sd <- aggregate(Coffee ~Fruit.Cup,cafe,sd)
coffee.mean <-  aggregate(Coffee ~Fruit.Cup,cafe,mean)
```
