```r
x <- 10
y <- runif(1) *100

if( x > y ){
    print("x > runif()")
} else{
     print("x <= runif()")
}

print("---end---")
```

```r
x <- 10
y <- runif(1) *100

if( x > y ){
    print("x > runif()")
} else if( x < y){
     print("x < runif()")
} else {
     print("x = runif()")
}
```

```r
# 홀수 짝수 구분
a <- 8
if (a %% 2 == 0) {
    sprintf("%i는 짝수입니다.",a)
} else if(a %%2 == 1) {
     sprintf("%i는 홀수입니다.",a)
}
```

```r
#여러 데이터를 한번에 받을때 ifelse
a <- c(8,7,6)   
ifelse(a %% 2 == 0,"짝수","홀수")
ifelse(a %% 2 == 0,sprintf("%i : 짝수",a),sprintf("%i : 홀수",a))
ifelse(a %% 2 == 0,paste(a,": 짝수"),paste(a,": 홀수"))
```

```r
# 사용자로부터 data를 입력 받아 입력한 숫자가 홀수인지 짝수인지 판단하여 출력하기 
num1 <- as.integer(readline("숫자를 입력하세요 :"))
# 입력하신 숫자는 12이고 짝수입니다.
if (num1 %% 2 == 0 ){
    sprintf("입력하신 숫자는 %i이고, 짝수입니다",num1)
} else {
    sprintf("입력하신 숫자는 %i이고, 홀수입니다",num1)
}
```

```r
# 점수에 따른 학점 입력받기 
num1 <- as.numeric(readline("점수를 입력하세요 :"))
if(num1 >=90) {
    print("A학점입니다.")
} else if(num1 >=80) {
    print("B학점입니다.")
} else if(num1 >=70) {
    print("C학점입니다.")
} else if(num1 >=60) {
    print("D학점입니다.")
} else {
    print("F학점입니다.")
}
```
