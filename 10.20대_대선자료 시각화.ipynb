```r
# 데이터 불러오기 
election <- read.csv("../Data/election_2022.csv")
election
# 선거인수 = 선거 가능한 인원
# 투표수 = 실제로 투표한 인원 
# 이,윤,심의 합이 투표수는 아님(나머지후보 투표, 무효수 등)
```

```r
summary(election)
```

```r
str(election)
```

```r
library(dplyr)                     #arrange 쓸 때
vote투표 <- aggregate(투표수 ~ 광역시도,election,sum)
vote투표 <- arrange(vote투표,desc(투표수))
vote투표$ratio <- round(    (vote투표$투표수 / sum(vote투표$투표수))*100 ,2)      #투표율이 100일때 경기도, 서울 등 
vote투표
label <- paste(vote투표$광역시도,"\n",vote투표$ratio,"%")
```

```r
par(family ="AppleGothic")
options(warn=-1)
pie3D(vote투표$ratio,
    main = "광역시도별 투표율",
    labels = label,
    labelcex = 0.8,
    radius = 0.8            #default =1
)
```

```r
# electionSum = 이재명 + 윤석열  + 심상정
election$electionSum <- apply((election)[,5:7],1,sum)
election
```

```r
# electionRatio <- 투표수 대비 electionSum의 퍼센트 구해서 넣기 
election$electionRatio <- round((election$electionSum /election$투표수)*100,2)
election
```

```r
#  3명의 후보를 가장 많이 선택한 지역과 가장 적게 선택한 지역 구하기 
# 1)정렬
library(dplyr)
arrange(election,desc(electionRatio))[1,]                   #max
arrange(election,desc(electionRatio))[nrow(election),]      #min
```

```r
# 2) subset
subset(election,electionRatio == max(election$electionRatio))
subset(election,electionRatio == min(election$electionRatio))
```

```r
# election_2022.jo1.csv로 저장하기 
write.csv(election,"../Data/election_2022.jo1.csv",row.names =F)
election
```

```r
#  각후보 (윤석열,이재명,심상정)의 광역시도별 득표율을 3D pie chart로 표현하기 
# 단, 득표울이 3%미만인 광역시는 광역시도 명만 출력하고 퍼센트는 출력허지 않음 , 페센트 단위는 
# 특정 득표율 = 투표율 / 전체 투표수 
library(plotrix)
```

```r
#심성정 득표율 표 만들기 
voteSim <- aggregate(심상정 ~ 광역시도,election,sum)                #광역시도별 심상정 투표수 
ratio <- round(voteSim$심상정/sum(voteSim$심상정)*100,1)
voteSim$투표율 <- ratio
voteSim <- arrange(voteSim, desc(투표율))
voteSim
```

```r
par(family = 'AppleGothic')
# options(warn = -1)
label <- paste(voteSim$광역시도,"\n", ifelse(voteSim$투표율 >=3, paste((voteSim$투표율),"%"),""))
pie3D(voteSim$투표율,
   main = "심상정 후보의 광역시도별 득표율",
    labels =label,
    labelcex = 0.8,
   )
```

```r
#윤석열
par(family = "AppleGothic")

윤석열 <- arrange(aggregate(윤석열~광역시도,election,sum),desc(aggregate(윤석열~광역시도,election,sum)[,2]))
윤석열$ratio <- (윤석열$윤석열 / sum(윤석열$윤석열))*100
pie3D(윤석열$ratio,
      main = "윤석열 후보의 광역시도별 득표율",
      labels = paste(윤석열$광역시도,"\n",
                     ifelse(윤석열$ratio>=3,paste(round(윤석열$ratio,1),"%"),"")),
      labelcex = 0.4)
```

```r
# 이재명 득표율
votes <- aggregate(이재명 ~ 광역시도,election,sum)
votes <- arrange(votes,desc(votes$이재명))

options(warn = -1)
투표수 <- votes$이재명
시군구 <- votes$광역시도
label <- paste(시군구,"\n", ifelse(votes$이재명/sum(votes$이재명)*100 >=3 ,paste(round(votes$이재명/sum(votes$이재명)*100,2),"%"),""))

par(family = 'AppleGothic')
pie3D(투표수,
      main = "이재명 광역시도별 투표율",
      labels = label,
      labelcex = 0.8,
#       radius = 0.3 (파이 작아짐)
    )
```

```r
# 수도권 (경기도,서울, 안천)의 투표인구수의 퍼센트 ? 
cat(round(sum(subset(election,광역시도 %in% c('경기도','서울특별시','인천광역시'))$투표수) / sum(election$투표수) *100,2),'%')
```
