```r
# 샘플 번호 따오기 
samp <- sample(1:nrow(Cclass_ML), round(nrow(Cclass_ML)*0.7))
```

```r
samp
```

```r
data.tr <- Cclass_ML[samp,]
data.test <- Cclass_ML[-samp,]
```

```r
head(data.tr)
```

```r
# 모델학습 
# 라이브러리 import 
library(caret) 
library(e1071) 
library(party) 
library(nnet) 
library(randomForest) 
library(Metrics)
```

```r
# data.tr의 문제, 정답 만들기 
data.tr.x <- data.tr[,-1]
data.tr.y <- data.tr[,1]
```

```r
# data.test의 문제, 정답 만들기 
data.test.x <- data.test[,-1]
data.test.y <- data.test[,1]
```

```r

# 모델 만들기 
# 다중선형회귀분석 
model_lm <- lm(price ~ ., data = data.tr)
model_lm
```

```r
summary(model_lm)
```

```r
# 트레이닝 데이터 성능 
pred_tr <- predict(model_lm, newdata = data.tr.x)
```

```r
# correct = 정답 / predicted = 예측 
correct <- data.tr.y
predicted <- pred_tr
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
mae(correct, predicted)
```

```r
# 테스트 데이터 성능 
pred_test <- predict(model_lm, newdata = data.test.x)
# correct = 정답 / predicted = 예측 
correct <- data.test.y
predicted <- pred_test
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
mae(correct, predicted)
```

> 0.01 차이
