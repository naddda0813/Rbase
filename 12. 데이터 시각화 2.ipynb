### 한 화면에 차트 여러개 그려서 비교하기 

```r
price <- runif(10,1,100)            #1부터 100까지의 난수 10개 생성 
plot(price)
```

```r
# 2행 2열 차트 그리기 :type 속성으로 구분해보기 
# 같은 셀 안에 있어야 함, 여러 모양으로 보여줄때 (막대,파이 가능)
par(mfrow = c(2,2))
plot(price, type='l') (위에 왼쪽)
plot(price, type='o')
plot(price, type='h') (아래왼쪽)
plot(price, type='s')
```

```r
# 2행 2열 차트 그리기 :pch 속성으로 구분해보기
par(mfrow =c(2,2))

plot(price, type='o',pch=5)
plot(price, type='o',pch=15)
plot(price, type='o',pch=20,col='blue')
plot(price, type='o',pch=c(rep(15,8),20,20),        #15번을 8번반복후, 20번 두개 
            col='orange',cex=c(rep(1,8),3,1)
            )
```

#   종합 자료 시각화

```r
x<- c(1,2,3,4,2,4)
y<- rep(2,6)
```

```r
plot(x,y)       # 중첩되어 4개만 보임 
```

```r
# Method 1)
xy.df<-as.data.frame(table(x,y))
xy.df
plot(x,y,
    pch = 15,
    cex= 2 * xy.df$Freq             # 빈도수가 많은 건 더 크게 
)
```

```r
# Method 2)중복된 데이터도 따로 보여주기 
x <- c(1,1,1,2,2,2,2,2,2,3,3,4,4,5,6)
y <- c(1,1,4,2,3,2,2,2,3,2,1,1,1,1,1)
z <- data.frame(x,y)
sunflowerplot(z)
```

```r
# iris의 Sepal.Length, Sepal.Width의 중복된 데이터 보기 
x <- iris$Sepal.Length
y <-iris$Sepal.Width
z <- data.frame(x,y)
sunflowerplot(z,
    cex = 0.8,
    col= 'orange'
)
```

```r
options(warn=-1)
par(family='AppleGothic')

#기타 그래프
x <- c(1, 3, 6, 8, 9)
y <- c(12,56,78,32,9)
plot(x,y)
# 선분 그리기 (3,3) 과 (4,4)
segments(6,78,8,32,col='red') #1번(x,y),2번(x,y)
# 화살표 그리기 # 순서를 바꾸면 역방향 
arrows(1,12,3,56,col='blue') #1번(x,y),2번(x,y)  
# 사각형 그리기 
rect(4,20,6,30,col="lightblue",border ='red',density= 10)       #시작점과 끝점, 대각선 기준
# 글자 쓰기 
text(4,40,"this is 샘플",col='white',srt=50)            #srt 기울기
#테두리 활용하기 
mtext("character",side=3, adj = 0.5)                        #별도의 라벨 
mtext("text",side=4, adj = 0.5)                        #별도의 라벨 
# chart의 테두리 그리기 
box(lty=1,col='red')
#  x축의 추가 : y = 40 위치에 0~ 10까지 표시하고 색은 노란색
axis(1,pos=40,at=0:10,col='yellow')                    #1방향 :x축
#  x축의 추가 : x = 6 위치에 20~ 60까지 표시하고 색은 초록색
axis(2,pos=6,at=20:60,col='green')                    #2방향 :y축
```

# 수학함수의 시각화 


```r
y <- x^2
```

```r
f <- function(x){
    y = x^2
    return(y)
}
```

```r
x<-1:3
f(x)
```

```r
x <- seq(-5,5,0.1)
plot(x,f(x),
    type="l"
)
```

```r
y = e^(-x) - 4x^2 + x +3(x)-3
```

```r
f <- function(x) { 
    y = exp(-x) - 4 * x^2 + x + 3 *abs(x) -3
}
```

```r
plot(x,f(x),
type="l"
)
```

```r
# y= x^2 + 2x + 3
f <- function(x) { 
    y= (x)^2 + 2 * x + 3
}
```

```r
plot(x,f(x),
type="l"
)
```
