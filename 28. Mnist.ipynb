- V1 은 정답,V1은 뺀 784(28 *28)은 한 컬럼에 한 칸

```r
# Data 내용 확인하기 
num1 = seq(2,785,28)    # 시작, 끝 ,더해지는 값
num2 = seq(29,785,28)
cnt = 0 
for(j in 1:28) {
    n1 = as.numeric(num1[j])
    n2 = as.numeric(num2[j])
        for(i in seq(n1,n2,1)) {        
        if(mnist[3,i] != 0){        #1에 숫자 넗기
            cat("*")
        } else{
            cat("  ")    #   구분
        }
        cnt = cnt +1
        if( cnt>27) {
            cat("\n")
            cnt = 0
        }
    }
}
```

```r
# v1 을 factor(변경 )
mnist$V1 <- factor(mnist$V1)
str(mnist)
```

```r
# sampling
library(caret)
set.seed(1234)
```

```r
samp <- createDataPartition(mnist$V1,p=0.7,list=F)
```

```r
#학습용 
data.tr <- mnist[samp,]
# 검증용
data.test <- mnist[-samp,]
```

```r
# 문제와 정답
x <- subset(data.test,select=-V1)
y <- data.test$V1
```

```r
library(nnet)
library(party)
library(e1071)
library(randomForest)
```
