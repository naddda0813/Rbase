```r
# 표준화 하기 (인공신경망)
Cclass_ML_scale <- as.data.frame(scale(Cclass_ML))
```

```r
# 가격만 표준화 안 하고 원래 가격 넣기 
Cclass_ML_scale$price <- Cclass_ML$price
```

```r
# 표준화 샘플 번호 
samp2 <- sample(1:nrow(Cclass_ML_scale), round(nrow(Cclass_ML_scale)*0.7))
```

```r
data.tr2 <- Cclass_ML_scale[samp2,]
data.test2 <- Cclass_ML_scale[-samp2,]
```

```r
head(data.tr)
```

```r
# 모델학습 
# 라이브러리 import 
library(caret) 
library(e1071) 
library(party) 
library(nnet) 
library(randomForest) 
library(Metrics)
```

```r
# data.tr의 문제, 정답 만들기 
data.tr.x <- data.tr[,-1]
data.tr.y <- data.tr[,1]
```

```r
# data.test의 문제, 정답 만들기 
data.test.x <- data.test[,-1]
data.test.y <- data.test[,1]
```

```r
# data.tr2의 문제, 정답 만들기 
data.tr2.x <- data.tr2[,-1]
data.tr2.y <- data.tr2[,1]
```

```r
# data.test2의 문제, 정답 만들기 
data.test2.x <- data.test2[,-1]
data.test2.y <- data.test2[,1]
```

```r
# 모델 만들기 
# 인공신경망 
model.nnet <- nnet(price ~.,data = data.tr2, size = 17, linout =T)
```

```r
# 트레이닝 2 데이터 성능 
pred_tr <- predict(model.nnet,newdata = data.tr2.x, type = "raw")
```

```r
# correct = 정답 / predicted = 예측 
correct <- data.tr2.y
predicted <- pred_tr
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
mae(correct, predicted)
```

```r
# 테스트 데이터 성능 
pred_test <- predict(model.nnet, newdata = data.test2.x)
# correct = 정답 / predicted = 예측 
correct <- data.test2.y
predicted <- pred_test
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
mae(correct, predicted)
```
