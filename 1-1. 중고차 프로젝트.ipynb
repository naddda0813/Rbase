## transmission 별 판매량 막대그래프

```r
head(benz5)
```

```r
benz5$transmission <- str_replace_all(benz5$transmission,"-","")
```

```r
# transmission 별 판매량
table(benz5$transmission)
```

```r
sum(table(benz5$transmission))
```

```r
length(table(benz5$transmission))
```

```r
#막대그래프
par(family="AppleGothic")
options(warn=-1)
barplot(table(benz5$transmission),
        main = "Benz transmission별 판매량",
        ylim = c(0,6000),
        xlab = "transmission",
        ylab= "판매대수"
       )
for(i in 1:length(table(benz5$transmission))) {
    text(0.7+ (i-1)*1.2,table(benz5$transmission)[i]/2,paste(table(benz5$transmission)[i],"대"))
}
```

```r
# transmission 별 판매비율 , 소수 둘째 자리에서 반올림 
Automatic<- round(nrow(subset(benz5,transmission == "Automatic")) / nrow(benz5) *100,2) ; Automatic
Manual<- round(nrow(subset(benz5,transmission == "Manual")) / nrow(benz5) *100,2) ;Manual
SemiAuto <- round(nrow(subset(benz5,transmission == "SemiAuto")) / nrow(benz5) *100,2);SemiAuto
```

```r
trans <- cbind(Automatic,Manual,SemiAuto)
trans
```

```r
# transmission 별 판매비율의 그래프로 바꾸기
par(family="AppleGothic")
options(warn=-1)
barplot(trans,
        main = "Benz transmission별 판매비율",
       ylim= c(0,100),
       xlab = "Transmission",
        ylab = "판매비율",
        col = c('yellow'),
)

for(i in 1:length(trans)) {
    text(0.7+ (i-1)*1.2,trans[i]/2,paste(trans[i],"%"))
}
```

```r
# 모델별 상위 5개 판매량
modeltable<- sort(table(benz5$model),decreasing = TRUE)
modeltable
```
