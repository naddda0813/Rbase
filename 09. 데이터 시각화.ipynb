### 서울시 교통사고 데이터로 발생건수와 부상자수의 산포도 구하기 

```r
seoul <- read.csv ("../Data/newSeoul2019.csv")
head(seoul)
```

```r
# 발생건수의 숫자가 150 이상 빨간색 ,미만이면 파란색  
par(family = "AppleGothic")

plot(seoul$발생건수, seoul$부상자수,
    main = '서울시 교통사고 발생건수와 부상자수 선포도',        
    xlab="발생건수",
    ylab = "부상자수",
    col = ifelse(seoul$발생건수 >= 150, "red",'blue')
)
```

```r
# main title : Simple Time Series
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
 xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
 ylim = c(min(y)-1, max(y)+1),
 main = 'Simple Time Series'   
 )
```

```r
# 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "l"                      #꺽은선 타입
 )
```

```r
# 점과 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "b"                      #점과 꺽은선 타입
 )
```

```r
# 점이 빠진 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "c"                      #점이 빠진 꺽은선 타입
 )
```

```r
# 점과 꺽은선이 중첩한 그래프
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "o",                       #점과 꺽은선이 중첩 
    col = c(rep('red',3),'blue',rep('green',6))
 )
```

```r
# 수직선 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "h",                       # 수직선
    col = c(rep('red',3),'blue',rep('green',6))
 )
```

```r
# 왼쪽 값을 기초로한 계단 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "s",                     
 )
```

```r
# 오른쪽 값을 기초로한 계단 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "S", 
    lty = 6            # line type                     
 )
```

```r
# 포인트에 심볼 주기 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "p", 
    pch= c(24,20,9),            # line type 
    col=c('red','maroon','yellow')                   
 )
```

```r
x <- runif(100)
y <- runif(100)
# pch : y > 0.5 이면  pch: 1, 아니면 18
# col : y > 0.5 이면  col: red, 아니면 blue
plot(x,y,
    col = ifelse(y > 0.5,'red','blue'),
    pch = ifelse(y > 0.5,1,18)
)
```
