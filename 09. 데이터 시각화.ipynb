```r
# 수직선 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "h",                       # 수직선
    col = c(rep('red',3),'blue',rep('green',6))
 )
```

```r
# 왼쪽 값을 기초로한 계단 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "s",                     
 )
```

```r
# 오른쪽 값을 기초로한 계단 그래프 
y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    xlim = c(min(y)-1, max(y)+1),  #최소값에서 1을 빼고, 최대값에서 1을 더한다. 
    ylim = c(min(y)-1, max(y)+1),
    main = 'Simple Time Series',
    type = "S", 
    lty = 6            # line type                     
 )
```

# 선그래프 (Line Graph)

```r
# 3종류 dataset의 선 그래프 
abc <-c(260,300,250,200,310)
def <- c(180,200,210,190,170)
ghi <- c(210,250,260,210,270)
```

```r
# plot 
par(family='AppleGothic')
plot(abc,
    type = "o" ,           #선과 점이 겹친 형태, overlap 
    col = "#ad5a23",
    ylim = c(100,400),        #limit가 없으면 x축과 y축의 데이터폭이 안 맞음 
    axes = F,               # x,y 좌표 Frame 제거 ,= f라 해야 제거됨(false)
    ann = F                 # x,y 의 label 제거 , = f라 해야 제거됨(false)
    )

# x축과  x축의 라벨을 설정 
axis(1,at=1:5, lab=c('서울','대전','대구','광주','원주'))       #1: x축, at:인덱스 범위,label 

#y축과  y축의 라벨을 설정
axis(2,ylim=c(100,400))

# main title 지정 ,plot 외부에서도 가능, 컬러값을 줄때 
title(main ="과일 판매량" , col.main="#276b2c")

# x축 title(label)지정
title(xlab ="지역",col.lab = "#6e4dab")

#  y축 title(label)지정
title(ylab ="판매량",col.lab = "#2b0a22" )

# 선그리기 함수로 def 그리기 
lines(def,                    #이어지는 선으로 그려짐 
    type = "o",                #인덱스값에 점 생긴
    col= "#5ead91",
    pch = 22,
    lty = 2                     #점선 
)                

# 선그리기 함수로 ghi 그리기 
lines(ghi,
    type = "o", 
    pch = 22,
    col = "#274f8f",
    lty =2                      #line type ,2 ,점선 
)
# 범례
legend(4, 400,          # x,y좌표 
        c("야구장","축구장","해변가"),
        col = c("#ad5a23","#5ead91","#274f8f"),
        pch = c(21,22,22),
        lty =c(1,2,2),
        cex = 0.8         #크기조정,  defalut = 1 , 1보다 작은 값일시 작게 보임 
)
```

```r
# V1 : 성별 (남,녀)
# V2 : 점수
# V3 : 과목 (국어 ,영어, 수학)
# V4 : 학력 (초,중,고,대학교,대학원)

data$V1 <- factor(data$V1, levels =1:2,labels = c('남','여'))           #int를 factor로 변환
data$V3 <- factor(data$V3, levels =1:3,labels = c('국어' ,'영어', '수학'))           #int를 factor로 변환
data$V4 <- factor(data$V4, levels =1:5,labels = c('초','중','고','대','원'))           #int를 factor로 변환
```
